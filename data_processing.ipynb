```python
import pandas as pd

# Load CSV file
df = pd.read_csv('Airline_Delay_Cause.csv')  # Replace with your actual file path
```

```python
# Display the first few rows
print(df.head())
```

```python
# Display the descrpition of the dataset
df.describe()
```

```python
# PREPROCESSING 1: SPLIT AIRPORT NAME
# Splitting city, state, and airport name
df[['airport_city', 'airport_state']] = df['airport_name'].str.split(', ', expand=True)
df[['airport_state', 'airport_name']] = df['airport_state'].str.split(': ', expand=True)

# Add airport_city and airport_state to the DataFrame
df['airport_city'] = df['airport_city'].str.strip()
df['airport_state'] = df['airport_state'].str.strip()
```

```python
# PREPROCESSING 2: CHANGE TO PERCENTAGE DATA

# Calculate percentages for delay 15 minutes or more
df['carrier_ct_percentage'] = (df['carrier_ct'] / df['arr_del15']) * 100
df['weather_ct_percentage'] = (df['weather_ct'] / df['arr_del15']) * 100
df['nas_ct_percentage'] = (df['nas_ct'] / df['arr_del15']) * 100
df['security_ct_percentage'] = (df['security_ct'] / df['arr_del15']) * 100
df['late_aircraft_ct_percentage'] = (df['late_aircraft_ct'] / df['arr_del15']) * 100

# Calculate percentages for total delay minutes
df['carrier_delay_percentage'] = (df['carrier_delay'] / df['arr_delay']) * 100
df['nas_delay_percentage'] = (df['nas_delay'] / df['arr_delay']) * 100
df['weather_delay_percentage'] = (df['weather_delay'] / df['arr_delay']) * 100
df['security_delay_percentage'] = (df['security_delay'] / df['arr_delay']) * 100
df['late_aircraft_delay_percentage'] = (df['late_aircraft_delay'] / df['arr_delay']) * 100

# Add all percentage columns to the DataFrame (for delay 15 minutes or more)
df['carrier_ct_percentage'] = df['carrier_ct_percentage'].round(2)
df['weather_ct_percentage'] = df['weather_ct_percentage'].round(2)
df['nas_ct_percentage'] = df['nas_ct_percentage'].round(2)
df['security_ct_percentage'] = df['security_ct_percentage'].round(2)
df['late_aircraft_ct_percentage'] = df['late_aircraft_ct_percentage'].round(2)

# Add all percentage columns to the DataFrame (for total delay minutes)
df['carrier_delay_percentage'] = df['carrier_delay_percentage'].round(2)
df['nas_delay_percentage'] = df['nas_delay_percentage'].round(2)
df['weather_delay_percentage'] = df['weather_delay_percentage'].round(2)
df['security_delay_percentage'] = df['security_delay_percentage'].round(2)
df['late_aircraft_delay_percentage'] = df['late_aircraft_delay_percentage'].round(2)
```

```python
# Reorder columns
ordered_columns = [
    'year', 'month', 'carrier', 'carrier_name', 'airport',
    'airport_city', 'airport_state', 'airport_name',  # Preprocessing 1 result
    'arr_flights', 
    'arr_del15', 
    'carrier_ct', 'carrier_ct_percentage',
    'weather_ct', 'weather_ct_percentage',
    'nas_ct', 'nas_ct_percentage',
    'security_ct', 'security_ct_percentage',
    'late_aircraft_ct', 'late_aircraft_ct_percentage',
    'arr_cancelled', 'arr_diverted', 
    'arr_delay',
    'carrier_delay', 'carrier_delay_percentage', 
    'weather_delay', 'weather_delay_percentage',
    'nas_delay', 'nas_delay_percentage',
    'security_delay', 'security_delay_percentage',
    'late_aircraft_delay', 'late_aircraft_delay_percentage'
]

# Reorder the DataFrame columns
df = df[ordered_columns]

# Display the first few rows of the modified DataFrame
print(df.head())
```

```python
# Save to a new CSV file
df.to_csv('Airline_Delay_Cause_Data_Processing.csv', index=False)
print("New CSV file 'Airline_Delay_Cause_Data_Processing.csv' saved successfully!")
```
